Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Importing data from csv file

```python
df = pd.read_csv('C:/Users/d/Desktop/uw/hw/w7/data2.csv')
df
```

Getting average score for movie

```python
ma = df.groupby(['movie'])
ma.mean()
```

Getting average score for person

```python
pa = df.groupby(['person'])
pa.mean()
```

Creating normalized dataframe 

```python
df = df.assign(normalized=(df.score - df.score.min())/(df.score.max() - df.score.min()))
```

Comparing normalized and original averages

```python
pa = df.groupby(['person'])
pa.mean()
```

```python
ma = df.groupby(['movie'])
ma.mean()
```

Conclusion: Using normalized data lets you see results knowing the range, which may help to analyse data. Witouth normalizing data and not knowing maximum value it might be tricky to get a correct view of data, for example if we would be looking at number of items bought in one year time, the max value migth be unknown for us and when we would analyze data by month we will not have clear reference. The disadvantage is that we lose meta data information, for example if we would see only normalized data we would not know that our movie rating system is 1-5, which may affect our judgement since scores would probablly be different if we would had 1-100 system. Both data types are useful and are better used together than alone.
